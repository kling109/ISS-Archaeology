# Data Processing for the ISS Archaeology Project

```python
import json
import matplotlib.pyplot as plt
import networkx as nx
import random
import numpy as np
```

```python
with open('rawdata.json') as jsonfile:
    data = json.load(jsonfile)
```

```python
data
```

```python
pairs = {}
for astros in data.values():
    allNames = list(astros.keys())
    for name1 in range(len(allNames)):
        for name2 in range(len(allNames))[name1+1:]:
            if str((allNames[name1], allNames[name2])) in pairs:
                pairs[str((allNames[name1], allNames[name2]))] += 1
            elif str((allNames[name2], allNames[name1])) in pairs:
                pairs[str((allNames[name2], allNames[name1]))] += 1
            else:
                pairs[str((allNames[name1], allNames[name2]))] = 1
```

```python
with open('pairs.json', 'w') as write_file:
    json.dump(pairs, write_file)
```

```python
USA_BASE = (100, 100) 
RUSSIA_BASE = (2200, 2200) 
JPN_BASE = (1200, 2000) 
GER_BASE = (1800, 1200) 
ITALY_BASE = (600, 1700) 
FRANCE_BASE = (1200, 2000) 
CAN_BASE = (2000, 200) 
BEL_BASE = (800, 1800)
KAZ_BASE = (400, 1500)
UK_BASE = (1000, 300)
BRAS_BASE = (1200, 500) 
```

```python
names = {}
for astros in data.values():
    for v in astros.keys():
        name = v.split('&')[0]
        nation = v.split('&')[1]
        if nation == 'canda':
            nation = 'canada'
        if nation not in names:
            names[nation] = [name]
        else:
            if name not in names[nation]:
                names[nation].append(name)
```

```python
# specify params
def genPoints(n):
    shape = np.array([300, 300])
    sensitivity = 0.5 # 0 means no movement, 1 means max distance is init_dist

    # compute grid shape based on number of points
    width_ratio = shape[1] / shape[0]
    num_y = np.int32(np.sqrt(n / width_ratio)) + 1
    num_x = np.int32(n / num_y) + 1

    # create regularly spaced neurons
    x = np.linspace(0., shape[1]-1, num_x, dtype=np.float32)
    y = np.linspace(0., shape[0]-1, num_y, dtype=np.float32)
    coords = np.stack(np.meshgrid(x, y), -1).reshape(-1,2)

    # compute spacing
    init_dist = np.min((x[1]-x[0], y[1]-y[0]))
    min_dist = init_dist * (1 - sensitivity)

    assert init_dist >= min_dist
    print(min_dist)

    # perturb points
    max_movement = (init_dist - min_dist)/2
    noise = np.random.uniform(
        low=-max_movement,
        high=max_movement,
        size=(len(coords), 2))
    coords += noise
    
    return coords
```

```python
usa = len(names["usa"])
points = genPoints(usa)
scatter = [p + USA_BASE for p in points]
scatter
```
